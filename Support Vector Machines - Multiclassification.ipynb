Parameter C is the inverse of the strength of regularization (penalty).   It also controls the trade off between a smooth decision boundary and classifying the training points correctly. So a high value of C will tell the model to choose a smaller-margin and does a better job of getting all the training points classified correctly.  A snaller C value would cause the model to create a larger seperating margin and allow more missclassification.  As the value of c increases, the model overfits, as it descreases, the model underfits. So its important to complete a gridsearch to find the optimal value for this parameter.

e) Describe conceptually what the purpose of a kernel is for Support Vector Machines.


When datasets are not linearly seperable, kernels become necessary to make effective decision boundaries between classes.  One approach to fixing this would be to add more features such as polynomial features. But with a high number of features or complex datasets, this becomes very computationally expensive. A kernel is capable of transforming the dataset into a higher dimensional space and providing a similar result without the computation expense.
